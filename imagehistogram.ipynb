```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/Fadia.JPG", cv2.IMREAD_UNCHANGED)

cv2_imshow(img)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
import numpy as np
a = np.array([[1,2,2,2,10],[10,20,30,40,50]])
print (a)
print (a.ravel())
print (a.flatten())
plt.hist(a.ravel (),bins = 50, range = [0,50])
plt.show()
```

```python
plt.hist(img.ravel(), bins = 256, range = [0, 256])
plt.show()
```

```python
color = ('b','g','r')
for i, col in enumerate(color):
  histogr = cv2.calcHist ([img], [i], None, [256], [0,256])
  plt.plot(histogr, color = col)
  plt.xlim([0,256])
  plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color) :
  histogr = cv2.calcHist ([img], [i], None, [4], [0,256])
  plt.plot(histogr, color = col)
  plt.xlim([0,4])
  plt.show()
```
